```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importando e verificando os dados
df = pd.read_csv('StudentsPerformance.csv')
df.head(10)
```

### Vendo se há números nulos

```python
plt.figure(figsize=(16,5))
sns.heatmap(df.isnull(), cbar = False);
```

### Gêreno

```python
gender = df['gender'].value_counts( normalize= True )

plt.figure(figsize=(4,6))
plt.bar(gender.index, gender, color = 'orange')
```

```python
plt.figure(figsize=(16,5))
plt.title("Percentual de Gênero")
plt.pie(gender, 
        labels=gender.index, 
        shadow= False, 
        startangle=90, 
        autopct='%1.1f%%');
```

```python
plt.figure(figsize=(12,5))
l_education = df['parental level of education'].value_counts()
plt.bar(l_education.index, l_education, color = 'orange');
```

### Escolaridade Parental

```python
plt.figure(figsize=(16,5))
plt.title("Percentual por Titulação")
plt.pie(l_education, 
        labels=l_education.index, 
        shadow= False, 
        startangle=90, 
        autopct='%1.1f%%');
```

### Situação alimentícia

```python
lunch = df['lunch'].value_counts( normalize= True )

plt.figure(figsize=(16,5))
plt.title("Percentual de Alimentação")
plt.pie(lunch, 
        labels=lunch.index, 
        shadow= False, 
        startangle=90, 
        autopct='%1.1f%%');
```

## Preparatório

```python
preparatorio = df['test preparation course'].value_counts(normalize=True)

plt.title("Percentual de quem fez preparatório")
plt.bar(preparatorio.index, preparatorio, color = 'orange');
```

```python
plt.figure(figsize=(16,5))
plt.title("Percentual por Preparação")
plt.pie(preparatorio, 
        labels=preparatorio.index, 
        shadow= False, 
        startangle=90, 
        autopct='%1.1f%%');
```
